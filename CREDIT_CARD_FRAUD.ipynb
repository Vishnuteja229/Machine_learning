```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
```

```python
card_data=pd.read_csv("..\\CSV files\\creditcard.csv")
```

```python
card_data
```

```python
card_data['Class'].unique()
```

```python
fraud_cases=card_data[card_data['Class']==1]
valid_cases=card_data[card_data['Class']==0]
```

```python
print(f"Fraud :{len(fraud_cases)} valid :{len(valid_cases)}")
```

```python
outlier=float(len(fraud_cases)/len(valid_cases))
print(outlier)
```

```python
cor_matrix=card_data.corr()
plt.figure(figsize=(8,8))
sns.heatmap(cor_matrix,vmax=.8,square=True,cmap='Set1')
plt.show()
```

```python
X=card_data.drop(['Class'],axis=1)
Y=card_data['Class']
```

```python
print(X.shape ,Y.shape)
```

```python
#X_data=X.values
#Y_data=Y.values
```

```python
(X_train,X_test,Y_train,Y_test)=train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
print(X_train.shape,Y_train.shape)
```

```python
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
Y_predict=model.predict(X_test)
```

# EVALUATING

```python
from sklearn.metrics import classification_report, accuracy_score  
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score, matthews_corrcoef 
from sklearn.metrics import confusion_matrix 
  
```

```python
print(f"Accuracy:{accuracy_score(Y_test,Y_predict)*100}")
print(f"Precision score:{precision_score(Y_test,Y_predict)*100}")
print(f"f1_score:{f1_score(Y_test,Y_predict)*100}")
print(f"matthews_corrcoef:{matthews_corrcoef(Y_test,Y_predict)*100}")
```

```python
c_repo=classification_report(Y_test,Y_predict)
print(c_repo)
```

```python
con_mat=confusion_matrix(Y_test,Y_predict)
plt.figure(figsize=(6,6))
sns.heatmap(con_mat,annot=True,fmt='d',xticklabels=['Valid','Fraud'],yticklabels=['valid','Fraud'])
plt.show()
```
